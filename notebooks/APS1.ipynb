```python
with open("./atp_dataset.csv") as file:
    filtered_lines = []
    reader = csv.reader(file, delimiter=',', quotechar='"')
    i = 0
    for line in reader:
        if i > 0:
            dates = line[3]
            year = dates.split("/")[2]
            if year == "15":
                filtered_lines.append(line)
        i+=1
print(filtered_lines)
```

```python
#filtering player lists
players1 = []
players2 = []
for i in filtered_lines:
    players1.append(i[15])
    players2.append(i[17])
```

```python
#getting list of players together but checking if there's a player repeated on the lists
players = []
for i in players1:
    if i not in players:
        players.append(i)

for j in players2:
    if j not in players:
        players.append(j)
```
